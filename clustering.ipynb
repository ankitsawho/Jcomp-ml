```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('phishing.csv')
```

```python
data.head()
```

```python
y = data['Result']
y
```

```python
data.drop('Result', axis=1,inplace=True)
```

```python
data
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(data)
```

```python
np.unique(kmeans.labels_)
```

```python
data['Labels'] = kmeans.labels_
```

```python
data['Results'] = y
```

```python
data
```

```python
data_0 = data.loc[data['Labels'] == 0]
data_1 = data.loc[data['Labels'] == 1]
data_2 = data.loc[data['Labels'] == 2]
```

```python
data_0.drop('Labels', axis=1, inplace=True)
data_1.drop('Labels', axis=1, inplace=True)
data_2.drop('Labels', axis=1, inplace=True)
```

```python
data_0.to_csv('./data/data_0.csv', index=False)
data_1.to_csv('./data/data_1.csv', index=False)
data_2.to_csv('./data/data_2.csv', index=False)
```

```python
import pickle 
with open('clustering_pickle', 'wb') as f:
    pickle.dump(kmeans, f)
```
